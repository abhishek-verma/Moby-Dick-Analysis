```python
from matplotlib import pyplot as plt
import numpy as np
import powerlaw as pwl
```

# Getting the data

```python
with open('words.txt') as f:
  word_freq = [ int(i) for i in f ]

index = list(range(0, len((word_freq))))

print(word_freq[0:5])
print(len(word_freq))
```

# Initial Plot Verifies long tail property of the dataset

```python
plt.plot(word_freq)
plt.show()
```

# Visualizing Scale free property

```python
plt.subplot(2, 2, 1)
plt.plot(index[:400], word_freq[:400])
plt.subplot(2, 2, 2)
plt.plot(index[400:4000], word_freq[400:4000])
plt.subplot(2, 2, 3)
plt.plot(index[4000:8000], word_freq[4000:8000])
plt.subplot(2, 2, 4)
plt.plot(index[8000:], word_freq[8000:])
```

# Powerlaw Analysis

```python
fit = pwl.Fit(word_freq)
```

### Calculating best minimal value for power law fit



```python
fit.power_law.alpha
```

### Log Log plot

```python
pwl.plot_pdf(word_freq)
```
